By "released title" we mean that the movie (or TV show) must have been shown publicly at least once (including festival screening).

- Users can update their votes as often as they'd like, but any new vote on the same title will overwrite the previous one, so it is one vote per title per user.

- IMDb title pages also include a Metacritic Score for a title, as well as user reviews and links to professional critic reviews from newspapers, magazines and other publications. They aim to offer a variety of opinions on a title so users can make informed viewing decisions. We also always display the breakdown of the ratings so users can see the distribution of votes and determine how uniform or polarized the opinion of a movie is.

# Function to make frequency bar graph for categorical data

```
def plotBarChart(data=None, column=None, xlabel=None):
  """Creates bar chart depicting frequency distribution of categorical features."""

  # Create a figure size of 15 X 7 inches.
  figure = plt.figure(figsize=[15, 7])

  # Extract labels and values using .value_counts() function.
  x_data = data[column].value_counts().index
  print("x data",x_data)
  y_data = data[column].value_counts().values
  print("Y data",y_data)

  # Using seaborn .bar() function
  sns.barplot(x=x_data, y=y_data, color='#1B9E77')

  # Set x, y labels of size 14 and title label of size 16.
  plt.xlabel(xlabel=xlabel, size=14)
  plt.ylabel(ylabel='Frequency', size=14)
  plt.title(label='Frequency Distribution of '+column, size=16)

  # Display the output
  plt.show()
```

# Function to make Pi Chart

```
def plotPieChart(data=None, column=None):
  """Creates a pie chart of categories"""

  # Create a figure size of 15 X 7 inches.
  figure = plt.figure(figsize=[15,10])

  # Seeting searbon style in pyplot interface
  plt.style.use('seaborn')
  sns.set_style("whitegrid")
  sns.set_color_codes("muted")

  # Extract labels and values using .value_counts() function.
  extract = data[column].value_counts().sort_index()
  labels, sizes = extract.index, extract.values

  # Initiate an explode list of categories
  explode_list = np.ones(len(data[column].unique()))/10

  # Using .pie() function to pass the xaxis and yaxis data
  plt.pie(x=sizes, labels=labels, explode=explode_list, autopct='%3.1f%%',startangle=180, shadow=True, wedgeprops=dict(width=0.20))
    
  plt.axis('equal')

  # Set title label of size 16.
  plt.title(label='Pie Distribution of Number of Movies in each '+column, size=16)

  # Display the output
  plt.show()
```

```
def plotHistChart(data=None, column=None, bins=10, xlabel=None):
  """Creates histogram chart depicting frequency distribution of numerical features."""

  # Create a figure size of 15 X 7 inches.
  figure = plt.figure(figsize=[15, 7])

  # Using seaborn .histplot() function
  sns.histplot(x=data[column], bins=bins, color='red', alpha=0.6)

  # Set x, y labels of size 14 and title label of size 16.
  plt.xlabel(xlabel=xlabel, size=14)
  plt.ylabel(ylabel='Frequency', size=14)
  plt.title(label='Frequency Distribution of movies with '+column, size=16)

  # Display the output
  plt.show()
```

## 8.7 What is the correlation between various variables?

## Scatter Plot between 2 Variables

```
def plotScatterChart(data=None, xaxis=None, yaxis=None):
  """Creates scatter chart of two numerical features."""

  # Create a figure size of 15 X 7 inches.
  figure = plt.figure(figsize=[15, 7])

  # Using seaborn .scatterplot() function
  sns.scatterplot(x=xaxis, y=yaxis, data=data, sizes=200, alpha=0.6, color='red')

  # Set x, y labels of size 14 and title label of size 16.
  plt.xlabel(xlabel=xaxis, size=14)
  plt.ylabel(ylabel=yaxis, size=14)
  plt.title(label='Scatter Distribution of '+xaxis+' & '+yaxis, size=16)

  # Display the output
  plt.show()
```

```
plotScatterChart(data=data, xaxis='Rating', yaxis='Metascore')
```
